# https://www.instagram.com/mi.sssssun/?hl=ko

```python
df1 = pd.DataFrame(columns=['date', 'src', 'tag'])

url = 'https://www.instagram.com/mi.sssssun/?hl=ko'

driver = webdriver.Chrome(path)

time.sleep(2)
driver.implicitly_wait(4)
driver.get(url)

total = driver.find_element_by_class_name("g47SY ").text
total = total.replace(",", "")
total = int(total)

time.sleep(5)
driver.find_element_by_class_name("_9AhH0").click() #click photo

for i in range(total):
    time.sleep(2)
    driver.implicitly_wait(4)
    
    date = driver.find_element_by_class_name("k_Q0X.NnvRN").find_element_by_tag_name("time").get_attribute("datetime")[0:10]
    date = datetime.strptime(date, '%Y-%m-%d')
    print('date1: ', date)

    if date > from_d:
        time.sleep(5)
        driver.implicitly_wait(4)
        driver.find_element_by_class_name('HBoOv.coreSpriteRightPaginationArrow').click() #next btn
        
        time.sleep(2)
        driver.implicitly_wait(4)   
        print('date2: ', date)
        
    elif to_d <= date <= from_d:
        time.sleep(5)
        print('date3: ', date) 

        try:
            time.sleep(5)
            driver.implicitly_wait(4)
            tag = driver.find_element_by_class_name("ZyFrc").find_element_by_tag_name("img").get_attribute("alt")
            pn = driver.find_element_by_class_name("ZyFrc").find_element_by_tag_name("img").get_attribute("src")
            print(tag)
            print(pn)
            time.sleep(2)
            driver.implicitly_wait(4)             

        except NoSuchElementException:
            pass

        except TypeError:
            pass    

        time.sleep(2)
        
        if ('사람' in tag) & ('근접' not in tag) & ('텍스트' not in tag) & ('음식' not in tag):
            series = pd.Series([date, url, pn, tag], index=['date', 'acc', 'src', 'tag']) 
            df1 = df1.append(series, ignore_index=True)
        
        time.sleep(5)
        driver.implicitly_wait(4)      

        driver.find_element_by_class_name('HBoOv.coreSpriteRightPaginationArrow').click() # next btn

    else:
        print('끝!!!!!!!')
        driver.close()
        break

end = time.time()
print('Running time: ', end - start)
```

```python
df1['지역'] = '서울'
```

```python
df1.to_csv('C:/njikim/pythonwork/weather-baby-clothes/df1.csv', encoding='utf-8', index=False)
```
